## NSTI distribution

```python
nsti = pd.read_csv('../input_data/marker_predicted_and_nsti.tsv',
                   sep='\t', header=0, index_col=0 )
nsti
```

```python
table = Artifact.load('../input_data/table.qza')
df = table.view(pd.DataFrame)
df
```

```python
# asv 974a1adf60f501f75cfbeae7714b9792 not in nsti table
set(df.columns) - set(nsti.index) 
```

```python
df = df.drop(columns=['974a1adf60f501f75cfbeae7714b9792'])
```

```python
asv2nsti = dict(zip(nsti.index, nsti['metadata_NSTI']))
```

```python
upper_bound = 0.01
lower_bound = 0
bottom=0
width = 0.75

cmap = mpl.cm.gist_ncar_r

fig, ax = plt.subplots(figsize=(6,5))

# https://matplotlib.org/mpl_toolkits/axes_grid/users/overview.html
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad=0.1)

while upper_bound < 2:    
    filt = [True if upper_bound>=asv2nsti[i]>lower_bound else False for i in df.columns ]
    nsti_percent = df.loc[:,filt].sum(axis=1)/df.sum(axis=1)*100
    ax.bar(df.index, nsti_percent, bottom=bottom, width=width, color=cmap(upper_bound/2))
    
    upper_bound+=0.01
    lower_bound+=0.01
    bottom+=nsti_percent

filt = [True if asv2nsti[i]>2 else False for i in df.columns ]
nsti_percent = df.loc[:,filt].sum(axis=1)/df.sum(axis=1)*100
ax.bar(df.index, nsti_percent, bottom=bottom, width=width, color='k')


cmap = mpl.cm.gist_ncar_r
norm = mpl.colors.Normalize(vmin=0, vmax=2)
fig.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap),
             cax=cax, orientation='vertical')

ax.set_axisbelow(True)
ax.grid(axis='y',c='lightgray', ls=':')
ax.set_ylabel('Cumulative ASV counts (%)', fontsize=12)
ax.set_xlim(-0.5-(1-width)/2, len(df.index)-0.5+(1-width)/2)
ax.set_xticklabels(df.index, rotation=90, va='top')

cax.set_ylabel('NSTI', fontsize=12)
cax.locator_params(nbins=4)


plt.tight_layout()
plt.savefig('../figures/NSTI distribution.png', dpi=600)
plt.show()    
```
